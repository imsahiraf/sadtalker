```python
# üó£Ô∏è Step 4: Convert text to audio using pyttsx3 (or you can upload your own)
!pip install TTS==0.22.0 numpy==1.24.4 --force-reinstall --no-cache-dir --quiet

from TTS.api import TTS

# Initialize (it will auto-download the model)
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False, gpu=False)

# Input text
text = "Welcome to the AI News. Today we cover top headlines in technology."

# Save to file
tts.tts_to_file(text=text, file_path="input.wav")
```
